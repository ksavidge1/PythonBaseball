# Fangraphs Batter Data - 2023 Season

We can use `pybaseball` to scrape the data off of the Fangraphs website for a desired season and choose between hitters, pitchers, or fielders. Let's look at just one for now.

```python
import pybaseball as pyb
fg_batters_2023 = pyb.batting_stats(2023)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Regression using numpy
WAR_per_player = fg_batters_2023['WAR']
wrcPlus_per_player = fg_batters_2023['wRC+']
m, b = np.polyfit(WAR_per_player, wrcPlus_per_player, 1)

# Scatterplot using matplotlib
fig, ax = plt.subplots(1,1, figsize=(8,8))
fig.set_facecolor("white")
ax.scatter(WAR_per_player, wrcPlus_per_player)
ax.plot(WAR_per_player, m*WAR_per_player + b)
ax.set_xlabel("WAR (Wins above replacement)")
ax.set_ylabel("wRC+ (Weighted Runs Created Plus)")
ax.spines[['right', 'top']].set_visible(False)

# Add some helpful indicators for wRC+ = 100 and WAR = 0, 
# as these are relevant baseline values given how each metric is calculated
ax.axhline(100, color = 'r', linestyle = '--')
ax.axvline(0, color = 'r', linestyle = '--')

plt.show()
```

```python
# seaborn example
import seaborn as sns

sns.displot(data=fg_batters_2023, x="Age", kde=True)
plt.show()
```

```python
# try adding a col argument, but restrict the range of Age for simpler display

fg_batters_2023_age_restricted = fg_batters_2023[np.logical_and(fg_batters_2023["Age"] >= 25,
                                                                      fg_batters_2023["Age"] <= 29)]
sns.displot(data=fg_batters_2023_age_restricted, x="WAR", col="Age", kde=True)
plt.show()
```
